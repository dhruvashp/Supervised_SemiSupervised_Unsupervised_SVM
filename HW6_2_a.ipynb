```python
import pandas as pd
import numpy as np

df = pd.read_csv('data_banknote_authentication.txt', header=None)
print(df)
```

```python
from sklearn.model_selection import train_test_split

bank_train,bank_test = train_test_split(df,shuffle=True,train_size=900)

print(bank_train)
print(bank_test)
```

```python
bank_train.reset_index(drop=True,inplace=True)
bank_test.reset_index(drop=True,inplace=True)
print(bank_train)
print(bank_test)
```

```python
bank_train.columns = ['X1','X2','X3','X4','y']
bank_test.columns = ['X1','X2','X3','X4','y']

print(bank_train)
print(bank_test)
```

```python
bank_train.to_csv('Bank_Train.csv')
bank_test.to_csv('Bank_Test.csv')

# To .csv files to use in next portions of the assignment
```
