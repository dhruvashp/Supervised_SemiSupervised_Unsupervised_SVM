Learning Curves obtained from Active and Passive Learning Errors

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

active = pd.read_csv('Active_Learning_Errors.csv',index_col=0)
passive = pd.read_csv('Passive_Learning_Errors.csv',index_col=0)

print(active)
print(passive)
```

```python
active_mean = active.mean(axis=0)
passive_mean = passive.mean(axis=0)
print(active_mean)
print(passive_mean)
```

```python
act = active_mean.to_numpy().flatten()
pas = passive_mean.to_numpy().flatten()

print(act)
print(pas)
```

```python
training_inst = np.array(active.columns).flatten().astype(int)
print(training_inst)
```

```python
plt.xlabel('Training Instances')
plt.ylabel('Test Errors')
red_patch = mpatches.Patch(color='red', label='Active')
blue_patch = mpatches.Patch(color='blue', label='Passive')
plt.legend(handles=[red_patch,blue_patch])
plt.plot(training_inst,act,'r',training_inst,pas,'b')
```

As is expected, and as can be seen, the Active Learner learns more quickly than the Passive Learner.
The test errors for the active learner reduce more steeply than that for the passive learner

Also note, that, as the training instances increase, the two curves tend to coincide, implying that as more and more training instances are added to the pool, the test errors reduce similarly, irrespective of the learner being active or passive.

Nonetheless, the active learner performs better than the passive learner for smaller number of training instances and thus has a tendency of learning more quickly than the passive learner

Also, for a large number of training instances, whereby almost all the training points are utilized, the test errors obtained are the same, learner irrespective.
