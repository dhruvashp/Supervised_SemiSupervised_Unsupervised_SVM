Before performing the Monte-Carlo iterations, we will select the range for parameter C, penalty, first, according to the method specified in the footnote

Even though not really required for range selection of the penalty parameter, we will randomly split original dataset into train and test, with test containing 20 % positive and 20 % negative class, in general.

Then over the train, we will fit the model for different C values, obtain TRAINING ERROR for these different C values, check which C values generate errors above the 30 % threshold and use this threshold to select the upper and lower ranges of C

NOTE : 

We first split data randomly into train and test (test, with the upper 1,0 proportions)
Then we loop over C (penalty), fit, obtain the train error, check the threshold and select C

Thus, as seen above, test data isn't used at all. Ideally, we should, perform range selection in each Monte-Carlo iteration before using the selected range further for cross-validation, obviously this will be very expensive computationally. Thus range selection will be done only once.

ADDITIONALLY : 

If the range selection, is done once, then we should select it in an averaged 'sense'. In each Monte-Carlo iterations the train and test will differ, so there is no true way to select the 'range', whilst ensuring that this range is selected in an averaged sense without at all seeing the test data, as the test data in MC keeps changing.

So the averaging, without 'cross-validation', may be performed over a similar Monte-Carlo iteration which selects train and test randomly, obtains range over train, and 'averages' selected range over all Monte-Carlo iterations.

This also is computationally expensive, and even if performed in a modular sense, time-consuming.


THUS, in conclusion, range selection will be done only ONCE, over a SINGLE randomly generated train/test set, without averaging (without Monte-Carlo averaging) to simplify computations

```python
norm_whole.sort_values(by=['y'],inplace=True)
print(norm_whole)
```

```python
norm_whole.reset_index(drop=True,inplace=True)
print(norm_whole)
```

```python
X_y_0 = norm_whole.iloc[0:357,:]
X_y_1 = norm_whole.iloc[357:569,:]
X_y_1.reset_index(drop=True,inplace=True)
print(X_y_0)
print(X_y_1)
```

```python
from sklearn.model_selection import train_test_split

X_y_train_0,X_y_test_0 = train_test_split(X_y_0,test_size = 0.2,shuffle=True)
X_y_train_1,X_y_test_1 = train_test_split(X_y_1,test_size = 0.2,shuffle=True)

print(X_y_train_0)
print(X_y_train_1)
print(X_y_test_0)
print(X_y_test_1)
```

```python
X_y_train = pd.concat([X_y_train_0,X_y_train_1],axis=0)
X_y_test = pd.concat([X_y_test_0,X_y_test_1],axis=0)
print(X_y_train)
print(X_y_test)
```

```python
from sklearn.utils import shuffle

X_y_train = shuffle(X_y_train)
X_y_test = shuffle(X_y_test)

print(X_y_train)
print(X_y_test)
```

```python
X_y_train.reset_index(drop=True,inplace=True)
X_y_test.reset_index(drop=True,inplace=True)
print(X_y_train)
print(X_y_test)
```

For Penalty Range Selection

We iterate over following very small and very large values of C, over the training set, obtained as above, and as such the test set for the range selection will not be utilized as we select parameters only over 'train'. This has been explained before.

Additionally the small and large values over which we iterate is as follows : 

C = {10^-6,10^-5,10^-4,10^-3,10^-2,10^2,10^3,10^4,10^5,10^6,10^7}

Training error threshold for selection = 30 %

Obviously L1 Penalty are used with Linear SVM's most popularly over normalized/standardized datasets. Thus here we will use L1 Penalty with Linear SVM

```python
X_train = X_y_train.drop(columns=['y'])
X_test = X_y_test.drop(columns=['y'])
y_train = X_y_train['y'].astype(int)
y_test =  X_y_test['y'].astype(int)
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

```python
from sklearn.svm import LinearSVC


C = np.array([10**(-6),10**(-5),10**(-4),10**(-3),10**(-2),10**(-1),10**(2),10**(3),10**(4),10**(5),10**(6),10**(7)])
error_df = pd.DataFrame(index=C,columns=['Error for Penalty Index'])

for i in np.arange(0,C.size):
    clf = LinearSVC(penalty='l1',dual=False,C=C[i],max_iter=100000).fit(X_train,y_train)
    y_train_pred = clf.predict(X_train) # Everything on Train for range selection
    
    mis = 0
    for j in np.arange(0,X_train.shape[0]):
        if y_train_pred[j] != y_train[j]:
            mis = mis + 1
    
    error_percent = (mis/X_train.shape[0])*100
    
    error_df.iloc[i,0] = error_percent
    
print(error_df)
```

These are errors, in percentage, over training data
As can be seen from the above dataframe, range for C for Supervised learning will be :

10^-1 (Lower Range)
10^+7 (Upper Range)

5-fold CV will be done over this range, increasing in a log fashion

Additionally, for C = 10^-2 we get error > 30 % or accuracy < 70 %, which is clearly above our error/below our accuracy threshold.

As such code could be written to select this, however as the selection was direct, easy and over a very small set of C, this was done directly by observation

NOTE : This was repeated quite a few times (thus with different random trains) and each time this range was obtained, thus we can call this the 'averaged overall' range, thus we don't even need to repeat this procedure as discussed previously due to the consistency of this range
